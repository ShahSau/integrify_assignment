### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
con=(euro_cap["Team"]=="England") | (euro_cap["Team"]=="Italy") | (euro_cap["Team"]=="Russia")
```

```python
s=euro_cap.loc[con,"Shooting Accuracy"]
print("Shooting Accuracy for England is : "+ s.iloc[0])
print()
print("Shooting Accuracy for Italy is : "+ s.iloc[1])
print()
print("Shooting Accuracy for Russia is : "+ s.iloc[2])
```
